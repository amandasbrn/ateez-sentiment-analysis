```python
hasilAnalisis = []
tweets_sentiment = []
for tweets_detail in hasilSearch:
    #if 'RT @' not in tweets_detail.text:
    '''
    data = {'Account name': ['@'+tweets_detail.user.screen_name], 
            'Tweet': [tweets_detail.text], 
            'Time': [tweets_detail.created_at]}
    tweets.append(data)
    '''
    tweet_properties = {}
    tweet_properties["tweet_date"] = tweets_detail.created_at
    tweet_properties["user"] = tweets_detail.user.screen_name
    tweet_properties["tweet"] = tweets_detail.text
    #bersihin tweet nya dari @user, hashtag dll
    tweet_bersih = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",tweets_detail.text).split())
        
    # print(tweet_bersih)
    
    # ngasih tau tipe data
    analysis = TextBlob(tweet_bersih)
    # print(type(analysis))
    
    # polarity = untuk ngehitung apakah sentimennya positif/netral/negatif
    # >0 = positif ; <0 = negatif ; 0 = netral
    # print(analysis.sentiment.polarity)
    if analysis.sentiment.polarity > 0.0:
        tweet_properties["sentiment"] = "positive"
    elif analysis.sentiment.polarity == 0.0:
        tweet_properties["sentiment"] = "netral"
    else:
        tweet_properties["sentiment"] = "negative"
    
    #biar retweet ga duplikat
    #if tweets_detail.retweet_count > 0:
    if 'RT @' not in tweets_detail.text:
        if tweet_properties not in hasilAnalisis:
            hasilAnalisis.append(tweet_properties)
        else:
            hasilAnalisis.append(tweet_properties)
    tweets.append(tweet_properties)
```
